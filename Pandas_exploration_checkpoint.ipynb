### Question 5
Select your strategy to handle missing values, and tell us why you had made that choice.
- I chose the median because it is not sensitive to outliers and it is more robust than the mean

```python
missing_columns = ["counter_number", "reading_remarque"]

for column in missing_columns:
    #compute the median for the current column
    median = client_0_bills[column].median()
    client_0_bills[column]=client_0_bills[column].fillna(median)
```

```python
#checking if the missing values have been filled
client_0_bills.isnull().sum()
```
